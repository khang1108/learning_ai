```python
import numpy as np
import pandas as pd

from matplotlib import pyplot as plt
```

```python
x = np.arange(0, 10)
y = np.array([1, 3, 2, 5, 7, 8, 8, 9, 10, 12])

plt.plot(x, y, "o r")
plt.show()
```

```python
def estimate_coef(x, y):
    n = np.size(x)

    mean_x = np.mean(x)
    mean_y = np.mean(y)

    SS_xy = np.sum(y * x) - n * mean_x * mean_y
    SS_xx = np.sum(x * x) - n * mean_x * mean_x

    b_1 = SS_xy / SS_xx
    b_o = mean_y - b_1 * mean_x

    return (b_o, b_1)
```

```python
b = estimate_coef(x, y)

y_pre = b[0] + x * b[1]

plt.plot(x, y, "ob")
plt.plot(x, y_pre, "-r")

plt.show()
```
